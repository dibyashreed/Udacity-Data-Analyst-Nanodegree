```python
file_url = "https://d17h27t6h515a5.cloudfront.net/topher/2017/August/599fd2ad_image-predictions/image-predictions.tsv"

# We are opening this file in wb(write binary) mode. That is because response.content is in bytes and not text.
# It will get rendered to text once we load it into a pandas dataframe. 
response = requests.get(file_url)
with open ('image-predictions.tsv', 'wb') as file:
    file.write(response.content)
```

```python
# Query the Twitter API for each tweets json data using Python's tweepy library and store each tweet's entire set of JSON data in a file. 

consumer_key = 'YOUR CONSUMER KEY'
consumer_secret = 'YOUR CONSUMER SECRET'
access_token = 'YOUR ACCESS TOKEN'
access_secret = 'YOUR ACCESS SECRET'

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth)
```

```python
# List of error tweets
error_list = []

# List of tweets
df_list = []



# For loop for adding each available tweet json into the dataframe list- df_list

for tweet_id in twitter_archive['tweet_id']:
    try:
        tweet = api.get_status(tweet_id, tweet_mode='extended',
                               wait_on_rate_limit = True, wait_on_rate_limit_notify = True)._json 
        favorites = tweet['favorite_count'] # Specifies how many favorites that tweet had.
        retweets = tweet['retweet_count'] # Specifes the number of retweets the tweet had.
        user_followers = tweet['user']['followers_count'] # How many followers the user had
        user_favourites = tweet['user']['favourites_count'] # How many favorites the user had
        date_time = tweet['created_at'] # The date and time of the creation
        
        df_list.append({'tweet_id'  : int(tweet_id),
                        'favorites' : int(favorites),
                        'retweets'  : int(retweets),
                        'user_followers' : int(user_followers),
                        'user_favorites' : int(user_favorites),
                        'date_time' : pd.to_datetime(date_time)})
        
    except Exception as e:
        print(str(tweet_id)+ " _ " + str(e))
        error_list.append(tweet_id)
```
